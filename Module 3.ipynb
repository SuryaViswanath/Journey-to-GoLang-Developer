### Composite datatypes
Congregation of other datatypes to acheive the complex datatypes to solve any problem. Thesse datatypes are:

1. **Arrays**
   * Fixed length of elements which all belong to the same datatype
   * They are stored in a sequential order
   * Array is initialized to zero if it is int or float and "" if it is string 
   ```go
   var lengthy [5] int
   ```
2. **Array Literal**
    * An array predefined with values
    ```go
var x [5] int = [5]{1,2,3,4,5}
var x [5] int = [...]{1,2,3,4,5}

```
    * ... can help in letting the compiler understand the length of the array
 To iterate through the array we do as follows:
 ```go
for i, v range x{
    fmt.Printf("ind %d, val %d", i,v)
}
```
_note the range function returns 2 values( index, value at index) for iterating through the array

3. **Slices**
    * Slice is a piece of the array that already exists.
    * Slice unlike array it can change the size with max at length of the array it is dependent on
    * Slice contains a pointer pointing to the first index; length determining the length of the slice; capacity captures the actual length the slice could scale up to
    
    ```go
arr:= [...]string{"a","b","c","d","r","g"}
s1:= arr[1:3]
s2:= arr[2:5]
len(s1) //gives out the length of the slice
cap(s1) //returns the max capacity of the source array
```
> output: 3,6

To access the slices we do arrayName[index]
    
4. **Slice Literals**
    similar to that of the array literal but we don't mention the size of the slice, it assumes this to be slice and points directly to the slice
    
    ```go
sli := []int{1,2,3}
```

There is another way to perform slicing over the existing array, this is done using a function called **make()**. The make function can either be called 2/3 arguments 
* 2 arguments - type and length
* 3 arguments - type, length and capacity

There is another function to add values to the slice, it is **append()**. Unlike any language we cannot add value to an array, the slicing here helps in increasing the length of the slice
*note: using the append function helps us in increasing the size greater than the underlying array size*

5. **Hash Tables**
    * Contains key/value pairs, where key is unique and value is the associated to the key

    * Hash function is used to generate the key to put the value that is to be inserted.( this is not called explicitly)

    * The value can be accessed using key rather than the index of the value

Advantages:
* Faster lookup than lists
* Arbitary keys

Disadvantages:
* May have collisions

6. Maps
* Implementation of hash table
* We use make() to create a map

```go
var idmap map[string]int
idmap = make(map[string]int)

idmap:=map[string]int{"Joe":123}

idmap["Jane"] = 456 //adding a value
delete(idmap,"Joe") //deleting a value
id, p:= idmap["Jane"] // here p is the boolean returning if jane exists
len(idmap) // gives the length of the map
```


The hash tables implementation in GOLang is carried out using the map function and to iterate through the map is done as follows:
```go
for key,val:=range idmap{
    fmt.Println(key,val)
}
```
